Q2) Perform all linear algebra operation with Tensorflow.

```python
import tensorflow as tf

A = tf.constant([[1, 2], [3, 4]], dtype=tf.float32)
B = tf.constant([[5, 6], [7, 8]], dtype=tf.float32)

# Addition
print(A + B)

# Matrix multiplication
print(tf.matmul(A, B))

# Transpose
print(tf.transpose(A))

# Determinant
print(tf.linalg.det(A))

# Inverse
print(tf.linalg.inv(A))
```

Q3) Write a program to implement AND OR gates using Perceptron.

```python
import numpy as np

def perceptron(x, w, b):
    return 1 if np.dot(x, w) + b >= 0 else 0

# AND gate
w_and = [1, 1]
b_and = -1.5

# OR gate
w_or = [1, 1]
b_or = -0.5

inputs = [(0,0),(0,1),(1,0),(1,1)]

print("AND Gate")
for i in inputs:
    print(i, perceptron(i, w_and, b_and))

print("\nOR Gate")
for i in inputs:
    print(i, perceptron(i, w_or, b_or))
```

Q4) Implementation of XOR Problem using PyTorch Neural Network.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Data
X = torch.tensor([[0,0],[0,1],[1,0],[1,1]], dtype=torch.float32)
y = torch.tensor([[0],[1],[1],[0]], dtype=torch.float32)

# Model
model = nn.Sequential(
    nn.Linear(2, 4),
    nn.ReLU(),
    nn.Linear(4, 1),
    nn.Sigmoid()
)

criterion = nn.BCELoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# Training
for epoch in range(2000):
    optimizer.zero_grad()
    output = model(X)
    loss = criterion(output, y)
    loss.backward()
    optimizer.step()

# Testing
print("XOR Output:")
print(model(X).round())
```

Q5) Implement Simple below Neural Network to solve regression problem.

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Dummy data
X = torch.unsqueeze(torch.linspace(-10, 10, 100), dim=1)
y = X * 2 + 1

# Model
class SimpleNN(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(1, 2)
        self.output = nn.Linear(2, 1)

    def forward(self, x):
        x = torch.relu(self.hidden(x))
        return self.output(x)

model = SimpleNN()
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# Training
for epoch in range(1000):
    optimizer.zero_grad()
    pred = model(X)
    loss = criterion(pred, y)
    loss.backward()
    optimizer.step()

print("Training completed")
```
